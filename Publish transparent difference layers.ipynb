```python
import geemap.foliumap as geemap 
import ee 
import datapane as dp

TurkanaMap_ndvichange_trans = geemap.Map(center = [3.1, 35.6], zoom = 8, add_google_map = False)
```

```python
%store -r diff_total
%store -r diff_90
%store -r diff_94
%store -r diff_95
%store -r diff_98
%store -r diff_99
%store -r diff_14
%store -r diff_15
%store -r diff_16
%store -r diff_17
%store -r diff_18
%store -r diff_19
%store -r diff_20
```

```python
# Setting visualization parameters 

diff_params_trans = {'bands': ['nd'], 'palette': ['e81f1f', 'edb418', 'e8f1b3', 'baf380', '49cfe8'], 'min': -0.5, 'max': 0.5, 'opacity': 0.5}
```

```python
# Add change layers to map 

TurkanaMap_ndvichange_trans.addLayer(diff_total, diff_params_trans, '1989 - 2020')
TurkanaMap_ndvichange_trans.addLayer(diff_94, diff_params_trans, '1989-1994')
TurkanaMap_ndvichange_trans.addLayer(diff_95, diff_params_trans, '1994-95')
TurkanaMap_ndvichange_trans.addLayer(diff_98, diff_params_trans, '1995-98')
TurkanaMap_ndvichange_trans.addLayer(diff_99, diff_params_trans, '1998-99')
TurkanaMap_ndvichange_trans.addLayer(diff_14, diff_params_trans, '1999-2014')
TurkanaMap_ndvichange_trans.addLayer(diff_15, diff_params_trans, '2014-15')
TurkanaMap_ndvichange_trans.addLayer(diff_16, diff_params_trans, '2015-16')
TurkanaMap_ndvichange_trans.addLayer(diff_17, diff_params_trans, '2016-17')
TurkanaMap_ndvichange_trans.addLayer(diff_18, diff_params_trans, '2017-18')
TurkanaMap_ndvichange_trans.addLayer(diff_19, diff_params_trans, '2018-19')
TurkanaMap_ndvichange_trans.addLayer(diff_20, diff_params_trans, '2019-20')

TurkanaMap_ndvichange_trans.add_colorbar(colors = diff_params_trans['palette'], vmin = -0.5, vmax = 0.5, caption = 'NDVI Change')
TurkanaMap_ndvichange_trans.add_layer_control()
TurkanaMap_ndvichange_trans
```

```python
dp.enable_logging()
TurkanaMap_ndvichange_trans.publish(name = 'NDVI Change Between Sample Years - Transparent Layers', description = 'Pixel values represent change in NDVI between sample years')
```
